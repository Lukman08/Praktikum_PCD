```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('color.png')
plt.imshow(img)
plt.title("Original Image")
```

```python
imgrgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(imgrgb)
```

```python
red = imgrgb[:,:,0]
plt.subplot(131)
plt.imshow(red)
plt.title("Layer Red")
plt.subplot(132)
plt.imshow(imgrgb[:,:,1])
plt.title("Layer Green")
plt.subplot(133)
plt.imshow(imgrgb[:,:,2])
plt.title("Layer Blue")
```

```python
imghsv = cv2.cvtColor(imgrgb, cv2.COLOR_RGB2HSV)
plt.imshow(imghsv)
```

```python
plt.subplot(131)
plt.imshow(imghsv[:,:,0])
plt.title("Layer Hue")
plt.subplot(132)
plt.imshow(imghsv[:,:,1])
plt.title("Layer Sturation")
plt.subplot(133)
plt.imshow(imghsv[:,:,2])
plt.title("Layer Value")
```

```python
imghls = cv2.cvtColor(imgrgb, cv2.COLOR_RGB2HLS)
plt.imshow(imghls)
```

```python
plt.subplot(131)
plt.imshow(imghls[:,:,0])
plt.title("Layer Hue")
plt.subplot(132)
plt.imshow(imghls[:,:,1])
plt.title("Layer Sturation")
plt.subplot(133)
plt.imshow(imghls[:,:,2])
plt.title("Layer Value")
```
