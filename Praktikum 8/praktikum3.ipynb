```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = 'data/nemo.jpg'
citra = cv2.imread(file)
citra = cv2.cvtColor(citra, cv2.COLOR_BGR2RGB)

file2 = 'data/goldhill.png'
CitraGray = cv2.imread(file2,0)
image = [citra, CitraGray]

for i in range(2):
    if citra is None:
        print('Tidak dapat membaca berkas', file)
    else:
        nkali=0
        # Cek sebagai citra berwarna atau bukan
        if len(image[i].shape) > 2:
            nkali = 3
            plt.imshow(image[i])
        else:
            nkali = 1
            plt.imshow(image[i], cmap=plt.get_cmap('gray') )
        plt.title(f'Citra{i}')
        plt.show()
        warna = ('r', 'g', 'b')
        for kanal in range(nkali):
            histo = cv2.calcHist([image[i]], [kanal], None, [256],[0,256])
            plt.plot(histo, color = warna[kanal])
        plt.title('Histogram ')
        plt.show()
```

```python
citra = cv2.imread('data/gembala.jpg', 0)
ekual = cv2.equalizeHist(citra)
hasil = np.hstack((citra, ekual))

# Tampilkan hasilnya
plt.imshow(hasil, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
citra = cv2.imread('data/gembala.jpg', 0)
ekual = cv2.equalizeHist(citra)

# Tampilkan gambar asal dan hasil menggunakan subplot
plt.subplot(121)
plt.imshow(citra, cmap = plt.get_cmap('gray'))
plt.xticks([]), plt.yticks([])
plt.title('Citra semula')

plt.subplot(122)
plt.imshow(ekual, cmap = plt.get_cmap('gray'))

plt.xticks([]), plt.yticks([])
plt.title('Citra hasil')
plt.show()
```

```python
citra = cv2.imread('data/gembala.jpg', 0)
clahe = cv2.createCLAHE(clipLimit = 12, tileGridSize = (16, 16))
ekual = clahe.apply(citra)
hasil = np.hstack((citra, ekual))

# Tampilkan hasilnya
plt.subplot(121)
plt.imshow(citra, cmap = plt.get_cmap('gray'))
plt.xticks([]), plt.yticks([])
plt.title('Citra semula')

plt.subplot(122)
plt.imshow(ekual, cmap = plt.get_cmap('gray'))

plt.xticks([]), plt.yticks([])
plt.title('Citra hasil')
plt.show()
```

```python
citra = cv2.imread('data/taipei.jpg')
lab = cv2.cvtColor(citra, cv2.COLOR_BGR2LAB)
kanalLAB = np.array(cv2.split(lab))
print(type(kanalLAB))
print(kanalLAB[0])
kanalLAB[0] = cv2.equalizeHist(kanalLAB[0])
lab = cv2.merge(kanalLAB)
bgr = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)
hasil = np.hstack((citra, bgr))
# Tampilkan hasilnya
cv2.imshow('Hasil', hasil)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
