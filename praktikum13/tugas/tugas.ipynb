```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#### Filter Morphologi
##### Fungsi untuk load gambar (create)

```python
def load_img(file_path):
    img = cv2.imread(file_path)
    return img

# Contoh penggunaan fungsi load_img
img = load_img('zoro.jpg')
cv2.imshow('Gambar Zoro', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Fungsi untuk menampilkan gambar

```python
# fungsi untuk menampilkan gambar
def display_img(load_image):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
plt.imshow(img)
```

```python
display_img(img)
grayZOro = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(grayZOro,cmap='gray')
```

```python
# erosi
kernel = np.ones((5,5),np.uint8)
erosion1 = cv2.erode(img,kernel,iterations = 1)
display_img(erosion1)
#img = load_img()
kernel = np.ones((5,5),np.uint8)
erosion5 = cv2.erode(img,kernel,iterations = 4)
display_img(erosion5)
```

```python
dilation = cv2.dilate(erosion5,kernel,iterations = 1)
display_img(dilation)
dilation5 = cv2.dilate(dilation,kernel,iterations = 1)
display_img(dilation5)
```

```python
img = load_img('zoro.jpg')
display_img(img)
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
display_img(gradient)
```
