```python
import cv2
import matplotlib.pyplot as plt
import numpy as  np
```

```python
 # load gambar dari komputer
inp = cv2.imread('data/zoro.jpg')#perhatikan path dari gambar
cv2.imshow('gambar original', inp)#menampilkan gambar
plt.imshow(inp)
plt.title("Gambar ori BGR")
plt.show()

#convert BGR to RGB
inpRGB = cv2.cvtColor(inp, cv2.COLOR_BGR2RGB)
plt.imshow(inpRGB)
plt.title("Gambar ori RGB")
plt.show()

tombol = cv2.waitKey(0) & 0xFF #jika menggunakan 32 bit, hapus “& 0xFF”
cv2.destroyAllWindows() #closing all windows
```

```python
d = inp.shape
r = d[0]
c = d[1]
out = inp
for x in range(0, r):
    for y in range(0, c):
        red = 255 - inp[x,y,0]
        green = 255 - inp[x,y,1]
        blue = 255 - inp[x,y,2]
        out[x,y] = [red, green, blue]
cv2.imshow('gambar negatif', out)#menampilkan gambar
#cara lain menampilkan file multimedia
outRgb = cv2.cvtColor(out, cv2.COLOR_BGR2RGB)
plt.imshow(outRgb)
plt.title("Gambar Negatif")
plt.show()

tombol = cv2.waitKey(0) & 0xFF #jika menggunakan 32 bit, hapus “& 0xFF”
if tombol==ord('s'):
    cv2.imwrite('gambarNegatif.png', out)
cv2.destroyAllWindows() #closing all windows
```

```python
imgGray = cv2.cvtColor(inpRGB, cv2.COLOR_RGB2GRAY)
plt.imshow(imgGray, cmap="gray")
```

```python
d = imgGray.shape
r = d[0]
c = d[1]
out = np.zeros(d)
T=120 #threshold
for x in range(0, r):
    for y in range(0, c):
        if (imgGray[x,y]<T):
            out[x,y] = 0
        else:
            out[x,y] = 255

plt.imshow(out, cmap="gray")
plt.title("Gambar Threshold")
plt.show()

tombol = cv2.waitKey(0) & 0xFF #jika menggunakan 32 bit, hapus “& 0xFF”
if tombol==ord('s'):
    cv2.imwrite('gambarthreshold.png', out)
cv2.destroyAllWindows() #closing all windows
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('data/logic.bmp')
img2 = cv2.imread('data/kotak.png')

img1 = cv2.resize(img1, (img2.shape[1], img2.shape[0]))

result = cv2.bitwise_and(img1, img2)

plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Citra 1')

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Citra 2')

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title('AND')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('data/logic.bmp')
img2 = cv2.imread('data/kotak.png')

gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

not_img1 = cv2.bitwise_not(gray1)

not_img2 = cv2.bitwise_not(gray2)

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title('Citra 1 Asli')

axs[0, 1].imshow(cv2.cvtColor(not_img1, cv2.COLOR_GRAY2RGB))
axs[0, 1].set_title('Citra 1 Setelah Operasi OR')

axs[1, 0].imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
axs[1, 0].set_title('Citra 2 Asli')

axs[1, 1].imshow(cv2.cvtColor(not_img2, cv2.COLOR_GRAY2RGB))
axs[1, 1].set_title('OR')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('data/logic.bmp')
img2 = cv2.imread('data/kotak.png')

img1 = cv2.resize(img1, (400, 400))
img2 = cv2.resize(img2, (400, 400))

gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

_, thresh1 = cv2.threshold(gray1, 127, 255, cv2.THRESH_BINARY)
_, thresh2 = cv2.threshold(gray2, 127, 255, cv2.THRESH_BINARY)

result = cv2.bitwise_xor(thresh1, thresh2)

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Citra 1')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Citra 2')

plt.subplot(1, 3, 3)
plt.imshow(result, cmap='gray')
plt.title('XOR')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

Tx = 20
Ty = 50

A = cv2.imread('data/cameraman.tif', 0)

brs, kol = A.shape

M = np.float32([[1, 0, Tx], [0, 1, Ty]])

B = cv2.warpAffine(A, M, (kol, brs))

plt.subplot(121), plt.imshow(A, cmap='gray')
plt.title('Asli'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(B, cmap='gray')
plt.title('Hasil'), plt.xticks([]), plt.yticks([])
plt.show()
```
