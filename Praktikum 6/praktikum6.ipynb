```python
import cv2
import matplotlib.pyplot as plt
import numpy as  np
```

```python
 # load gambar dari komputer
inp = cv2.imread('sonic.jpg')#perhatikan path dari gambar
cv2.imshow('gambar original', inp)#menampilkan gambar
plt.imshow(inp)
plt.title("Gambar ori BGR")
plt.show()

#convert BGR to RGB
inpRGB = cv2.cvtColor(inp, cv2.COLOR_BGR2RGB)
plt.imshow(inpRGB)
plt.title("Gambar ori RGB")
plt.show()

tombol = cv2.waitKey(0) & 0xFF #jika menggunakan 32 bit, hapus “& 0xFF”
cv2.destroyAllWindows() #closing all windows
```

```python
d = inp.shape
r = d[0]
c = d[1]
out = inp
for x in range(0, r):
    for y in range(0, c):
        red = 255 - inp[x,y,0]
        green = 255 - inp[x,y,1]
        blue = 255 - inp[x,y,2]
        out[x,y] = [red, green, blue]
cv2.imshow('gambar negatif', out)#menampilkan gambar
#cara lain menampilkan file multimedia
outRgb = cv2.cvtColor(out, cv2.COLOR_BGR2RGB)
plt.imshow(outRgb)
plt.title("Gambar Negatif")
plt.show()

tombol = cv2.waitKey(0) & 0xFF #jika menggunakan 32 bit, hapus “& 0xFF”
if tombol==ord('s'):
    cv2.imwrite('gambarNegatif.png', out)
cv2.destroyAllWindows() #closing all windows
```

```python
imgGray = cv2.cvtColor(inpRGB, cv2.COLOR_RGB2GRAY)
plt.imshow(imgGray, cmap="gray")
```

```python
d = imgGray.shape
r = d[0]
c = d[1]
out = np.zeros(d)
T=120 #threshold
for x in range(0, r):
    for y in range(0, c):
        if (imgGray[x,y]<T):
            out[x,y] = 0
        else:
            out[x,y] = 255

plt.imshow(out, cmap="gray")
plt.title("Gambar Threshold")
plt.show()

tombol = cv2.waitKey(0) & 0xFF #jika menggunakan 32 bit, hapus “& 0xFF”
if tombol==ord('s'):
    cv2.imwrite('gambarthreshold.png', out)
cv2.destroyAllWindows() #closing all windows
```
